```python
import json

def line_to_dict(line):
    indent = line.count('-')
    name = line.replace('-', '').strip().replace('/', '')
    return indent, {"name": name, "sub_file": [], "sub_folder": []}

def add_to_structure(structure, element, indent):
    if not structure:
        structure.append((indent, element))
    elif indent > structure[-1][0]:
        structure[-1][1]["sub_folder"].append(element)
        structure.append((indent, element))
    elif indent == structure[-1][0]:
        structure[-2][1]["sub_folder"].append(element)
        structure.append((indent, element))
    else:
        while indent <= structure[-1][0]:
            structure.pop()
        structure[-1][1]["sub_folder"].append(element)
        structure.append((indent, element))

def parse_structure(structure_str):
    lines = [line for line in structure_str.split('\n') if line.strip() and line.strip().startswith('-')]
    structure = []
    for line in lines:
        # '#' 기호로 시작하는 주석을 제거합니다.
        if '#' in line:
            line = line[:line.index('#')]
        if "/" in line:
            indent, element = line_to_dict(line)
            add_to_structure(structure, element, indent)
        else:
            indent, name = line.count('-'), line.replace('-', '').strip()
            if structure and indent > structure[-1][0]:
                structure[-1][1]["sub_file"].append(name)
            elif structure and indent == structure[-1][0]:
                structure[-2][1]["sub_file"].append(name)
    return structure[0][1]

def save_to_json(project_name, structure):
    data = {
        "project_name": project_name,
        "file_architecture": {
            "folder": [structure]
        }
    }
    
    return data

structure_str = """
- 거래/
-- app/
--- main.py
--- models.py
--- views.py
--- utils.py
-- static/
--- css/
---- style.css
--- js/
---- script.js
-- templates/
--- index.html
--- login.html
--- signup.html
-- test/
--- test/
---- test/
----- test.txt # ㅋㅋㅋㅋㅋ
아진짜로
"""

structure = parse_structure(structure_str)
project_name = structure['name']
save_to_json(project_name, structure)
```

```python
import pygraphviz as pgv

def create_graph(graph, folder):
    for file in folder['sub_file']:
        graph.add_node(file)

    for sub_folder in folder['sub_folder']:
        folder_name = sub_folder['name']
        graph.add_node(folder_name)
        graph.add_edge(folder['name'], folder_name)
        create_graph(graph, sub_folder)

        for file in sub_folder['sub_file']:
            graph.add_node(file)
            graph.add_edge(folder_name, file)

# 주어진 파일 및 폴더 구조 데이터
file_architecture = {
    "folder": [
        {
            "name": "거래",
            "sub_file": [
                "config.py",
                "run.py"
            ],
            "sub_folder": [
                {
                    "name": "app",
                    "sub_file": [
                        "main.py",
                        "wallet.py",
                        "trade.py"
                    ],
                    "sub_folder": []
                },
                {
                    "name": "static",
                    "sub_file": [],
                    "sub_folder": [
                        {
                            "name": "css",
                            "sub_file": [
                                "style.css",
                                "main.css"
                            ],
                            "sub_folder": []
                        },
                        {
                            "name": "js",
                            "sub_file": [
                                "script.js"
                            ],
                            "sub_folder": []
                        }
                    ]
                },
                {
                    "name": "templates",
                    "sub_file": [
                        "index.html",
                        "login.html",
                        "register.html"
                    ],
                    "sub_folder": []
                }
            ]
        }
    ]
}

# 프로젝트 이름
project_name = "거래"

# 그래프 생성
graph = pgv.AGraph(directed=True)

# 프로젝트 이름으로 루트 노드 추가
graph.add_node(project_name)

# 프로그램 흐름도 그리기
create_graph(graph, file_architecture['folder'][0])

# 그래프 저장
graph.layout(prog='dot')  # 그래프 레이아웃 설정 (여기서는 'dot' 사용)
graph.draw('flowchart.png')  # 그래프 이미지 저장
```
